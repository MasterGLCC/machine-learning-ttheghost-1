```python
# Importation des bibliothèques nécessaires

# On va utiliser sklearn pour la régression linéaire, numpy pour les opérations mathématiques, matplotlib pour la visualisation et pandas pour la manipulation des données.
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Charger les données
data = pd.read_csv('../../datasets/SOCR-HeightWeight.csv')

# Préparer les données
X = np.array(data['Height(Inches)'].values).reshape(-1, 1)  # Variable indépendante
y = np.array(data['Weight(Pounds)'].values)  # Variable dépendante
```

```python
# Créer et entraîner le modèle de régression linéaire
# On crée une instance de LinearRegression et on l'entraîne avec les données X et y.
model = LinearRegression()
# La méthode fit() est utilisée pour entraîner le modèle en ajustant les coefficients de la régression linéaire aux données d'entraînement.
model.fit(X, y)
print("Entrainement terminé.")
```

## Visualiser les données et la ligne de régression.

```python
# Afficher les paramètres du modèle
# D'apprer la documentation de sklearn, le coefficient de la régression linéaire est stocké dans l'attribut coef_ et l'ordonnée à l'origine (intercept) est stockée dans l'attribut intercept_.
a = model.coef_[0]  # Coefficient de la régression linéaire
b = model.intercept_  # Ordonnée à l'origine
print(f"La pente (coefficient): {a}")
print(f"L'ordonnée à l'origine: {b}")
print(f"Équation de la ligne de régression: y = {a:.2f}x + {b:.2f}")


# Faire des prédictions
y_pred = model.predict(X)

# Visualiser les données et la ligne de régression
plt.scatter(X, y, color='blue', label='Données réelles')
plt.plot(X, y_pred, color='red', label='Ligne de régression')
plt.xlabel('Taille (Inches)')
plt.ylabel('Poids (Pounds)')
plt.title('Régression Linéaire : Taille vs Poids')
plt.legend()
plt.show()
```
