```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Generer les données pour une regression linéaire multiple de 2 features
X, y = make_regression(n_samples=1000, n_features=2, noise=10, random_state=42)
```

```python
# Créer le modèle de regression linéaire
model = LinearRegression()
# Entraîner le modèle sur les données
model.fit(X, y)
# Faire des prédictions
y_pred = model.predict(X)
```

```python
# Afficher les coefficients et l'intercept du modèle
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Utiliser matplotlib pour visualiser les données et la ligne de régression en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:, 0], X[:, 1], y, color='blue', label='Données réelles')
# Afficher les prédictions come un plan
ax.plot_trisurf(X[:, 0], X[:, 1], y_pred, color='red', alpha=0.5, label='Plan de régression')

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Target')

ax.legend()
plt.show()
```
